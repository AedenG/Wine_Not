```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sqlalchemy  
from sqlalchemy import create_engine 
```

```python
# Example python program to read data from a PostgreSQL table

# Create an engine instance
alchemyEngine  = create_engine('postgresql+psycopg2://postgres:password@localhost/Wine_Not_DB')

# Connect to PostgreSQL server
dbConnection   = alchemyEngine.connect();

# Read data from PostgreSQL database table and load into a DataFrame instance
dataFrame = pd.read_sql("select * from \"combined_wine\"", dbConnection);
pd.set_option('display.expand_frame_repr', False);

# Print the DataFrame
print(dataFrame);

# Close the database connection
dbConnection.close();
```

```python
combined_wine = dataFrame  
combined_wine.head()
```

```python
combined_wine
```

```python
X = combined_wine.drop(columns="wine_type")
y = combined_wine['wine_type']

print("Shape: ", X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)

```

```python
#Create the model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
   max_iter=1000,
   random_state=1)
```

```python
#Use the score() method on the model to show the R2 score
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
