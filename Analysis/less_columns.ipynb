```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random

red_df = pd.read_csv('../Resources/binary_red_wine_df.csv')
white_df = pd.read_csv('../Resources/binary_white_wine_df.csv')
```

```python
red_df
```

```python
white_wine = white_df.sample(n=1599, random_state=1)
white_wine
```

```python
combined_wine = pd.DataFrame.append(red_df, white_wine)
combined_wine
```

```python
white_wine = white_wine.sample(n=5, random_state=1)
white_wine
```

```python
X = combined_wine.drop(columns=["wine_type", "quality", "chlorides", "fixed acidity", "pH", "residual sugar", "citric acid", "density", "alcohol"])
y = combined_wine['wine_type']

print("Shape: ", X.shape, y.shape)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
#Create the model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
   max_iter=4000,
   random_state=1)
```

```python
#Use the score() method on the model to show the R2 score
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
